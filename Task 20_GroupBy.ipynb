# Q.1) Group the DataFrame by 'Department' and calculate the mean salary for each department.
* data = {'Employeel D': [101, 102, 103, 104, 105],
'Employee Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Department': ['HR', 'IT', 'HR', 'Finonlinance', 'IT'],
'Salary': [60000, 80000, 70000, 90000, 85000],
'Joining Date': [date(2018, 5, 15), date (2019, 2, 10), date(2018, 8, 22), date(2020, 1, 5), date(2019, 7,
12)]}
* Group the DataFrame by 'Department' and get the group with the highest average salary.


```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
```

```python
#DataFrame 
data = {'Employeel D': [101, 102, 103, 104, 105],
'Employee Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Department': ['HR', 'IT', 'HR', 'Finonlinance', 'IT'],
'Salary': [60000, 80000, 70000, 90000, 85000],
'Joining Date': [(2018, 5, 15),(2019, 2, 10),(2018, 8, 22),(2020, 1, 5),(2019, 7,12)]}

df = pd.DataFrame(data)
df
```

```python
#DataFrame by 'Department' and calculate the mean salary for each department.
df.groupby("Department")[["Salary"]].mean()
```

```python
# Group the DataFrame by 'Department' and get the group with the highest average salary.
df.groupby('Department')['Salary'].mean().idxmax()
```

* Create a new column 'YearsInCompany' that calculates the number of years each employee has been with the company. Group the DataFrame by 'Department'    and get the group with the employee who has been with the company for the longest time.

```python
# Group the DataFrame by 'Department' and create a new DataFrame that shows the count of employees in each department.(
df.groupby("Department")[["Employee Name"]].count().reset_index(names = "Employee Count")
```

```python
df["Joining Date"] = df["Joining Date"].apply(lambda x: pd.Timestamp(*x))
```

```python
df["Joining Date"] = pd.to_datetime(df["Joining Date"])
df["YearsInCompany"] = (pd.Timestamp.now() - df["Joining Date"]).dt.days // 365
```

```python
df
```

```python
# Group the DataFrame by 'Department' and get the group with the employee who has been with the company for the longest time.
df.groupby("Department")["YearsInCompany"].max().idxmax()
```
