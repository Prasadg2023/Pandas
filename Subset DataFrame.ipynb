# Filtering

```python
df_iris["SepalLengthCm"] > 5
```

```python
df_iris[df_iris["SepalLengthCm"] > 5]
```

```python
df_iris.loc[df_iris["SepalLengthCm"] > 5, "Species"]
```

# Based on two criteria with Operator

```python
df_iris[(df_iris["SepalLengthCm"] > 5) & (df_iris["SepalLengthCm"] < 6)]
```

```python
df_iris[df_iris["SepalLengthCm"].between(5,6)] #same as above
```

```python
df_iris[(df_iris["SepalLengthCm"] > 5) | (df_iris["SepalLengthCm"] < 6)] #with or operator
```
