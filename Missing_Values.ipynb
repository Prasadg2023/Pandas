# Handling Missing Values

```python
df
```

```python
df.dropna() #drop missing value rows
```

# Parameters in dropna that we can use

* axis
* how
* thresh
* subset
* inplace

```python
df.dropna(axis = 1)
```

# thresh


```python
df.dropna(axis = 1, thresh = 7)
```

```python
df.dropna(axis = 1, thresh = 7).count()
```

```python
df.dropna()
```

# how

```python
df.dropna(how = "any")
```

```python
df.dropna(how = "all")
```

# subset

```python
df
```

```python
df.dropna(subset = ["Name", "English"])
```

```python
df.dropna(subset = ["Name", "Science "])
```

# inplace

```python
df #iot is not permenant
```

```python
df.dropna(inplace = True) #it makes permanant
```

```python
df
```

# drop

```python
df
```

```python
df.drop("English", axis = 1) #drop enlish column
```

```python
df.drop(1) #drop first index row
```

# droping multiple column

```python
df.drop(["English", "Science "], axis = 1) #drop mentioned columns
```

```python
df.drop(columns = ["English", "Science "]) #same as above without mentioning axis
```

# Filling Missing Values

```python
df
```

```python
df.fillna(0, axis = 1) #fill 0 where nan values present
```

# Backward fill

```python
df.fillna(method = "bfill")
```

# Forward fill

```python
df.fillna(method = "ffill")
```

# Mean

```python
df
```

```python
df.dtypes
```

```python
df["English"].fillna(df["English"].mean())
```

```python
df["Science "].fillna(df["Science "].mean())
```

# Median

```python
df["English"].fillna(df["English"].median())
```

```python
df["Science "].fillna(df["Science "].median())
```
