# Data types in Pandas

```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
#creating Data Frame
data = {"Car": ["BMW", "ALTO", "AUDI","MARUTI", "HONDA", "HYUNDAI", "FORD", "KIA"],\
     "Color": ["Red", "Yellow", "Black", "Green", "Black", "Red", "Black", "Black"],\
     "Year": ["1990", "1980", "2003", "2000", "2001", "2004", "1999", "2020"],\
     "Rating": ["2.5", "1.5", "3.8", "9.7", "8.9", "3.2", "5.5", "6.9"],\
     "Service": ["30/04/2023", "31/08/1999", "28/12/2015", "29/06/2011",\
                 "30/12/2020", "31/07/2013", "28/11/2000", "25/12/2020"]}
df = pd.DataFrame(data)
df.head()
```

```python
df.dtypes
```

# Converting the data types of columns

```python
df["Year"].astype("int")
```

```python
df["Year"] = df["Year"].astype("int")
```

```python
df.dtypes
```

```python
df["Rating"] = df["Rating"].astype("float")
```

```python
data = {"Car": ["BMW", "ALTO", "AUDI","MARUTI", "HONDA", "HYUNDAI", "FORD", "KIA"],\
     "Color": ["Red", "Yellow", "Black", "Green", "Black", "Red", "Black", "Black"],\
     "Year": ["1990", "1980", "2003", "2000", "2001", "2004", "1999", "2020"],\
     "Rating": ["2.5", "1.5", "3.8", "9.7", "8.9", "3.2", "5.5", "6.9"],\
     "Service": ["30/04/2023", "31/08/1999", "28/12/2015", "29/06/2011",\
                 "30/12/2020", "31/07/2013", "28/11/2000", "25/12/2020"]}
df1 = pd.DataFrame(data)
df1
```

```python
df1.dtypes
```

```python
#using pd.to_numeric
df1["Year"] = pd.to_numeric(df1["Year"])
```

```python
df1.dtypes
```

```python
df1["Rating"] = pd.to_numeric(df1["Rating"])
```

```python
df1.dtypes
```

```python
df1["Service"] = pd.to_datetime(df1["Service"])
```

```python
df1.dtypes
```
