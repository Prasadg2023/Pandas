```python
import pandas as pd

dict1={ 
       "City":["Simla","Delhi","Jaipur","Bhopal"],
       "Temp":[15,33,38,35]
}

dict2={ 
       "City":["Chennai","Hyderabad","Vizag","Kochi"],
       "Humid":[39,34,36,30]
}

dfnorth=pd.DataFrame(dict1)
dfsouth=pd.DataFrame(dict2)
```

```python
display(dfnorth,dfsouth)
```

```python
pd.concat([dfnorth,dfsouth])
```

```python
pd.concat([dfnorth,dfsouth],ignore_index=True)
```

```python
df_key=pd.concat([dfnorth,dfsouth],keys=["North","South"])
df_key
```

```python
df_key.loc["North"]
```

```python
df_key.loc["North"]["Temp"]
```

# Joins

```python
dict5={ 
       "City":["Chennai","Hyderabad","Vizag","Kochi","Mumbai"],
       "State":["TN","TS","AP","KS","MH"],
       "Temp":[15,33,38,35,30]
}

dict6={ 
       "City":["Chennai","Hyderabad","Vizag","Kochi","Pune"],
       "State":["TN","TS","AP","KS","MH"],
       "Humid":[39,34,36,30,39]
}

dftemp=pd.DataFrame(dict5)
dfhumid=pd.DataFrame(dict6)
```

```python
display(dftemp,dfhumid)
```

```python
pd.merge(dftemp,dfhumid)
```

```python
pd.merge(dftemp,dfhumid, on ="City")
```

```python
display(dftemp,dfhumid)
```

```python
pd.merge(dftemp,dfhumid, how ="inner")
```

```python
pd.merge(dftemp,dfhumid, how ="left")
```

```python
pd.merge(dftemp,dfhumid, how ="right")
```

```python
pd.merge(dftemp,dfhumid, how ="outer")
```

```python
pd.merge(dftemp,dfhumid, how ="cross")
```

```python
dict7={ 
       "City":["Simla","Delhi","Jaipur","Bhopal"],
       "Temp":[15,33,38,35]
}

dict8={ 
       "City":["Chennai","Hyderabad","Vizag","Kochi"],
       "Humid":[39,34,36,30]
}

df1=pd.DataFrame(dict7)
df2=pd.DataFrame(dict8)
```

```python
display(df1,df2)
```

```python
df1.join(df2,rsuffix="_R")
```

```python
df1.join(df2,lsuffix="_L")
```

```python
pd.concat([df1,dfsouth],axis=1)
```

```python
pd.concat([df1,df2],axis=0)
```
