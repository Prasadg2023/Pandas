# Date Range

```python
#day wise
print(pd.date_range("2000/1/11", periods = 5))
```

```python
#MOnth wise
print(pd.date_range("2000/1/11", periods = 5, freq = "M"))
```

```python
#Year wise
print(pd.date_range("2000/1/11", periods = 5, freq = "Y"))
```

```python
#create a datetime with specified intervals
start = pd.to_datetime("2011/1/1")
print("Start:",start)

end = pd.to_datetime("2011/1/20")
print("Start:",end)
print(pd.date_range(start,end))
```

```python
#creating dataframe with date column
df = pd.DataFrame({"name": ["Pk","RJ","YP"],
                   "DoB": ["01-02-2000","03-04-2001","05-06-2002"]})
```
