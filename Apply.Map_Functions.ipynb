```python
import pandas as pd
```

```python
df = pd.read_csv(r"data\Titanic.csv")
df
```

```python
gender = []
for i in df["Sex"]:
    if i == "male":
        gender.append(1)
    elif i == "female":
        gender.append(0)
df["Gender"] = gender
```

```python
df
```

# Map Function

```python
df["Gender_Map"] = df["Sex"].map({"male" : 1 ,"female": 0})
```

```python
df
```

# With lambda

```python
df["Gender_Lambda"] = df["Sex"].map(lambda x: 1 if x == "male" else 0)
```

```python
df
```

```python
df["Gender_apply"] = df["Sex"].apply(lambda x: 1 if x == "male" else 0)
```

```python
df
```

# create a age Class column
* Age < 18 - Child
* 18 < Age < 60 Adult
* Age> 60 Senior Adult

```python
df["Age_Class"] = df["Age"].map(lambda x: "Child" if x < 18 else "Adult" if 18 < x < 60 else "Senior Adult")
```

```python
df
```

```python
df["Age_Class"] = df["Age"].apply(lambda x: "Child" if x < 18 else "Adult" if 18 < x < 60 else "Senior Adult")
```
