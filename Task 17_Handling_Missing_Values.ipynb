# Q.1) You have a DataFrame student_data with columns 'StudentID', 'Name', 'Age', and 'Grade'. It contains missing values.
data = {'StudentID': [1, 2, 3, 4, 5],
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Age': [25, 22, None, 28, 30],
'Grade': [85, 92, 78, None, 95]}
* Perform the following tasks:
    * Display the count of missing values in each column.
    * Fill missing values in the 'Age' column with the mean age.
    * Remove rows with missing values in any column.

```python
import pandas as pd
#dataframe
data = {'StudentID': [1, 2, 3, 4, 5],
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Age': [25, 22, None, 28, 30],
'Grade': [85, 92, 78, None, 95]}

df = pd.DataFrame(data)
df
```

```python
# Display the count of missing values in each column.
df.isna().sum()
```

```python
# Fill missing values in the 'Age' column with the mean age.
df["Age"].fillna(df["Age"].mean())
```

```python
# Remove rows with missing values in any column
df.dropna()
```

# Q.2) You have a DataFrame employee_data with columns 'EmployeeID', 'Department', and 'Salary'. Some 'Department' values are missing.
data = {'EmployeeID': [101, 102, 103, 104, 105],
'Department': ['HR', 'IT', None, 'IT', 'HR'],
'Salary': [60000, 75000, 55000, 80000, 65000]}
* Perform the following tasks:
    * Display the count of missing values in the 'Department' column.
    * Impute missing 'Department' values with the most frequent department.

```python
#DataFrame
data = {'EmployeeID': [101, 102, 103, 104, 105],
'Department': ['HR', 'IT', None, 'IT', 'HR'],
'Salary': [60000, 75000, 55000, 80000, 65000]}

df1 = pd.DataFrame(data)
df1
```

```python
# Display the count of missing values in the 'Department' column.
print(df1["Department"].isna().sum())
```

```python
# Impute missing 'Department' values with the most frequent department.
df1["Department"].mode()[0]
```

```python
df1.columns
```

```python
df1["Department"].fillna(df1["Department"].mode()[0])
df1
```

# Q.3)  You have a DataFrame stock_prices with columns 'Date' and 'Closing_Price'. It represents daily stock prices.
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-04', '2022-01-05'],
'Closing_Price': [100, None, 120, None]}
* Perform the following tasks:
    * Display the count of missing values in the 'Closing_Price' column.
    * Fill missing 'Closing_Price' values using forward fill method for time series data.

```python
#DataFrame
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-04', '2022-01-05'],
         'Closing_Price': [100, None, 120, None]}

df2 = pd.DataFrame(data)
df2
```

```python
# Display the count of missing values in the 'Closing_Price' column.
print(df2["Closing_Price"].isna().sum())
```

```python
# Fill missing 'Closing_Price' values using forward fill method for time series data.
df2.fillna(method = "ffill")
```
