# Q.1) Suppose you have two Pandas DataFrames, df1 and df2, with a common column 'ID'. Perform an inner join on these DataFrames and display the result.
* df1 = ['ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']} df2 = {'ID': [2, 3, 4], 'Age': [25, 30, 22]}


```python
import pandas as pd
```

```python
# Perform an inner join on these DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}) 
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

display(df1,df2)
```

```python
pd.merge(df1, df2,how = "inner")
```

# Q.2) Given two DataFrames, left_df and right_df, perform a left join and display the result.
* left_df = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
* right_df = {'ID': [2, 3, 4], 'Age': [25, 30, 22]}

```python
# left join and display the result.¶
left_df = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
right_df = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})
display(left_df,right_df)
```

```python
pd.merge(left_df, right_df,how = "left")
```

# Q.3) Create two DataFrames, df7 and df8, and perform a cross join. Display the result.
* df7 = {'Color': ['Red', 'Green', 'Blue']} df8 = {'Size': ['Small', 'Medium', 'Large']}

```python
# cross join
df7 = pd.DataFrame({'Color': ['Red', 'Green', 'Blue']})
df8 = pd.DataFrame({'Size': ['Small', 'Medium', 'Large']})
display(df7,df8)                   
```

```python
pd.merge(df7, df8,how = "cross")
```

# Q.4) Create two DataFrames, df3 and df4, and perform an outer join. Display the result.
* df3 = {'ID': [1, 2, 3], 'City': ['New York', 'San Francisco', 'Los Angeles']} df4 = ['ID': [2, 3, 4], 'Population': [8500000, 880000, 4000000]}

```python
# outer join
df3 = pd.DataFrame({'ID': [1, 2, 3], 'City': ['New York', 'San Francisco', 'Los Angeles']}) 
df4 = pd.DataFrame({'ID': [2, 3, 4], 'Population': [8500000, 880000, 4000000]})
display(df3,df4)
```

```python
pd.merge(df3, df4,how = "outer")
```

# Q.5) Given two DataFrames, df5 and df6, perform a right join and display the result.
* df5 = ['ID': [1, 2, 3], 'Product': ['Laptop', 'Mouse', 'Keyboard']} df6 = {'ID': [2, 3, 4], 'Price': [1200, 30, 50]}

```python
# right join
df5 = pd.DataFrame({'ID': [1, 2, 3], 'Product': ['Laptop', 'Mouse', 'Keyboard']})
df6 = pd.DataFrame({'ID': [2, 3, 4], 'Price': [1200, 30, 50]})
display(df5,df6)
```

```python
pd.merge(df5, df6,how = "right")
```
