# PANDAS Series

```python
import pandas as pd
import numpy as np
```

```python
#creating series using 1D array
arr = np.arange(10,51,10)
print(arr, type(arr))
print()
s = pd.Series(arr)
print(s, type(s))
```

```python
# creating a series with mutable index
s = pd.Series(arr,index = ["first","second","third","fourth","fifth"])
print(s, type(s))              
```

```python
#Creating series using list of elements
s = pd.Series([11,22,33,44,55])
print(s, type(s))
```

```python
#Creating a series using scalar values
s = pd.Series(99, index = [1,2,3,4,5])
print(s, type(s))
```

```python
s = pd.Series(99, index = list(range(1001)))
print(s, type(s))
```

```python
s = pd.Series(np.nan, index = list(range(5)))
print(s, type(s))
```

```python
s = pd.Series(np.inf, index = list(range(5)))
print(s, type(s))
```

```python
#Creating a Series using python dictionary
d = {"name":"MS DHONI","Team":"CSK","Role":"Captain","Jersey number":7}
s = pd.Series(d)
print(s, type(s))
```

# PANDAS Data Frame

```python
#creating datafram
#empty
df = pd.DataFrame()
print(df)
```

```python
#converting a pandas series in dataframe
df = pd.DataFrame(s)
print(df)
```

```python
#Creating Dataframe from 2D array
arr = np.arange(10,101,10).reshape(5,2)
print(arr)
df=pd.DataFrame(arr)
print(df)
```

```python
df
```

```python
#naming the coloumns
arr = np.arange(10,101,10).reshape(5,2)
print(arr)
df=pd.DataFrame(arr, columns = ["col1","col2"])
print(df)
```

```python
df
```

```python
# creating data frame from list dictionaries.
d = [{"name":"Rohit Sharma","team":"MI"} , {"name":"Virat Kohali","team":"RCB"}, {"name":"Hardik Pandya","team":"MI"}]
df = pd.DataFrame(d)
df
```

```python
# creating data frame from list of list
d = [["Rohit Sharma","MI"] , ["Virat Kohali","RCB"], ["Hardik Pandya","MI"]]
df = pd.DataFrame(d, columns = ["name","team"])
df
```
