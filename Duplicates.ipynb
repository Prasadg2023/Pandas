# Duplicates

```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
# Creating data Frame
dict1 = {"Gender": ["Male","Female","Male","Female","Male"],
         "Married": ["Yes","No","No","No","Yes"],
         "Loan_Status":["Yes","No","No","No","Yes"]
        }
df = pd.DataFrame(dict1)
df
```

```python
df.duplicated()  #give true where duplicates occurs
```

```python
df.drop_duplicates() #duplicate value droped
```

# Droping Duplicates from particular columns

```python
df
```

```python
df.drop_duplicates(["Married"]) 
```

```python
#with keep="first" - ,mark duplicates as true for the first one
duplicates_first = df.duplicated(keep = "first")
duplicates_first
```

```python
#with keep="last" - ,mark duplicates as true for the last one
duplicates_first = df.duplicated(keep = "last")
duplicates_first
```

```python
df
```

```python
# In drop_duplicates with keep = "first" -keep duplicates which occured first time and drop others
df1 = df.drop_duplicates(keep= "first") 
df1
```

```python
# In drop_duplicates with keep = "last" -keep duplicates which occured last time and drop others
df1 = df.drop_duplicates(keep= "last") 
df1
```

```python
df
```

```python
df.drop_duplicates(keep="first",subset=["Gender","Married"])
```

```python
#with subset
df.drop_duplicates(keep="first",subset=["Married","Loan_Status"])
```

```python
df.drop_duplicates(keep="last",subset=["Married","Loan_Status"])
```
