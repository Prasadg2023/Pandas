# Q.1) Create a Pandas DataFrame with the following data:
large_dict = {'ID': np.arange(1, 100001),
'Name': [f'Person_{i}' for i in range(1, 100001)],
'Age': np.random.randint(18, 65, size=100000)
'Salary': np.random.randint(40000, 120000, size=100000)
'City': np.random.choice(['New York', 'London', 'Tokyo', 'Paris'], size=100000)}
Convert the dataframe into both CSV & Excel files and read those files.

```python
import pandas as pd
import numpy as np
#Pandas DataFrame
large_dict = {'ID': np.arange(1, 100001),
              'Name': [f'Person_{i}' for i in range(1, 100001)],
              'Age': np.random.randint(18, 65, size=100000),
              'Salary': np.random.randint(40000, 120000, size=100000),
              'City': np.random.choice(['New York', 'London', 'Tokyo', 'Paris'], size=100000)}

df = pd.DataFrame(large_dict)
df
```

```python
#convert to csv
df.to_csv("large_dict.csv",index = False)
```

```python
#covert to excel
df.to_excel("large_dict.xlsx",index = False)
```

```python
#read csv
pd.read_csv("large_dict.csv")
```

```python
#eard ecxel
pd.read_excel("large_dict.xlsx")
```

# Q.3) Create a Pandas DataFrame with columns 'City', 'Temperature', and 'Humidity' containing the following data:
| City        | Temperature | Humidity |
|-------------|-------------|----------|
| New York    | 75          | 60       |
| Chicago     | 68          | 55       |
| Los Angeles | 85          | 45       |
| Miami       | 80          | 70       |
* Display the rows where the Temperature is above 70.
* Display the rows where the Humidity is below 60.
* Display city with temperature 68 and humidity 55.

```python
#Pandas DataFrame with columns 'City', 'Temperature', and 'Humidity' containing the above data
data = {"City":["New York","Chicago","Los Angeles","Miami"],
        "Temperature":[75,68,85,80],
        "Humidity":[60,55,45,70]}

df = pd.DataFrame(data)
df
```

```python
#Display the rows where the Temperature is above 70.
print(df[df["Temperature"] > 70])
```

```python
# Display the rows where the Humidity is below 60.
print(df[df["Humidity"] < 70])
```

```python
#Display city with temperature 68 and humidity 55.
print(df[(df['Temperature'] == 68) & (df['Humidity'] == 55)])
```
