# Q.1) Create a Pandas DataFrame with the following data:
	data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        		  'Age': [25, 30, 35, 40],
        		 'Salary': [50000, 60000, 75000, 90000]}
* Print the 'Name' column from the above dataframe. Also Print the first two rows of the DataFrame.

```python
import pandas as pd
#creating dataframe 
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        		  'Age': [25, 30, 35, 40],
        		 'Salary': [50000, 60000, 75000, 90000]
       }
df = pd.DataFrame(data)
df
```

# Q.2) Print the details of the person with the highest and lowest salary from the above dataframe.
* Calculate and print the average age and salary.
* Select and print the rows where Age is less than 35 and Salary is greater than 60000.

```python
#printit highest and lowest salary
df = pd.DataFrame(data)
df
```

```python
#printing highest and lowest salary
highest_salary = df[df['Salary'] == df['Salary'].max()]
highest_salary
```

```python
lowest_salary = df[df['Salary'] == df['Salary'].min()]
lowest_salary
```

```python
#avg age
avg_age = df['Age'].mean()
print(avg_age)
```

```python
#avg Salary
avg_salary = df['Salary'].mean()
print(avg_salary)
```

```python
#printing the rows where Age is less than 35 and Salary is greater than 60000.
n_rows = df[(df['Age'] < 35) & (df['Salary'] > 60000)]
n_rows
```

# Q.3) Create a Pandas DataFrame from the following dictionary:
 	data = {'City': ['New York', 'London', 'Tokyo', 'Paris'],
 'Population': [8398748, 8982256, 37435191, 2140526],
 'Area_km2': [468.9, 1572, 2191, 105.4]}
* Add a new column 'Population_Density' calculated as Population/Area_km2.  Print the modified DataFrame.

```python
#creating Pandas DataFrame
import pandas as pd
data = {'City': ['New York', 'London', 'Tokyo', 'Paris'],
 'Population': [8398748, 8982256, 37435191, 2140526],
 'Area_km2': [468.9, 1572, 2191, 105.4]}


df = pd.DataFrame(data)
df
```

```python
#Add a new column 'Population_Density' calculated as Population/Area_km2. Print the modified DataFrame
df['Population_Density'] = df['Population'] / df['Area_km2']
df
```
