# Using multiple columns for groupby

```python
df.groupby(by = ["Married","Gender"])[["ApplicantIncome","LoanAmount"]].mean()
```

```python
df.groupby(by = ["Married","Gender"]).mean(numeric_only = True)
```

```python
# max value by grouoing
df.groupby(by = ["Married"])[["ApplicantIncome"]].max()
```

# Compare Average Income wit rspect to Gender and Married

```python
##using crosstab
pd.crosstab(df["Gender"], df["Married"], values = df["ApplicantIncome"], aggfunc = "mean")
```

```python
##using groupby
df.groupby(by =["Married","Gender"])["ApplicantIncome"].mean()
```

# Pivot Table

```python
df
```

```python
#Pivot table on ApplicantIncome mean value based on gender
df.pivot_table(values = "ApplicantIncome", index = "Gender", aggfunc = "mean")
```

```python
# Pivot table on ApplicantINcome mean value based  on fender and married column
df.pivot_table(values = "ApplicantIncome", index = "Gender",columns = "Married",aggfunc = "mean")
```

```python
# Pivot table on ApplicantINcome count value based  on fender and married column
df.pivot_table(values = "ApplicantIncome", index = "Gender",columns = "Married",aggfunc = "count")
```
